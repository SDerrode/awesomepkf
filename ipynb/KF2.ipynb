# Modèle linéaire pour le Filtre de Kalman

Ce notebook définit le système linéaire suivant :

$$X_{n+1} = F X_n + T_{n+1},$$
$$Y_{n+1} = G X_{n+1} + U_{n+1},$$

où $T_{n+1} -> N(0,Q)$ et $U_{n+1} -> N(0,R)$ sont des bruits gaussiens.

```python
import numpy as np

# Paramètres du modèle
F = np.array([[0.9]]) # Matrice de transition
G = np.array([[1.0]]) # Matrice d'observation
Q = np.array([[0.1]]) # Covariance du bruit de processus
R = np.array([[0.5]]) # Covariance du bruit de mesure
N = 50 # Nombre de pas de temps

# Initialisation des vecteurs
X = np.zeros((1, N+1)) # Etat vrai
Y = np.zeros((1, N+1)) # Observations

# Simulation du système
for n in range(N):
    T = np.random.multivariate_normal(mean=[0], cov=Q)
    U = np.random.multivariate_normal(mean=[0], cov=R)
    X[:, n+1] = F @ X[:, n] + T
    Y[:, n+1] = G @ X[:, n+1] + U

# Affichage d'un aperçu
print('Premiers états X:', X[:, :5])
print('Premières observations Y:', Y[:, :5])
```
